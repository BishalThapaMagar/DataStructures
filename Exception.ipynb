# Syntax

```python
try:
    # risky code
except SomeException as e:
    # handle exception
else:
    # runs if no exception
finally:
    # always runs
```

# handle any BS exception

```python
try:
    lst = [1,2]
    print(lst[5])  # Raises IndexError automatically
except Exception as e:
    print(e)
```

# Check for exception by self / custom exception?

```python
def check_age(age):
    try:
        if age < 0:
            raise ValueError("Age cannot be negative!")  # manually raising exception
        print(f"Your age is {age}")
    except ValueError as e:
        print(f"Caught an exception: {e}")  # handling the exception here
    else:
        print("Exception free code runs only if no exception \ni.e. Output : ",age)  # runs if no exception was raised
    finally:
        print("Other code here, to run no matter what")  # runs always


check_age(25)
print("-----")
check_age(-5)
```
