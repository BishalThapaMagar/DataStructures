# Stack

```python
class Stack:
    def __init__(self):
        self.Stack=[]

    def push(self,item):
        self.Stack.append(item)

    def pop(self):
        if len(self.Stack)==0:
            raise IndexError()
        return self.Stack.pop() # pop[0] will pop first one, #insert(0,data) will intert in front for stack

    def display(self):
        print(self.Stack)
```

```python
Stack = Stack()
Stack.push(1)
Stack.push(2)
Stack.push(3)
Stack.pop()
Stack.display()
```

Queue

```python
class Deque:
    def __init__(self):
        self.items = []

    # Add to right end
    def append_right(self, x):
        self.items.append(x)

    # Add to left end
    def append_left(self, x):
        self.items.insert(0, x)

    # Remove from right end
    def pop_right(self):
        if not self.items:
            raise IndexError("pop from empty deque")
        return self.items.pop()

    # Remove from left end
    def pop_left(self):
        if not self.items:
            raise IndexError("pop from empty deque")
        return self.items.pop(0)

    # See leftmost element
    def front(self):
        if not self.items:
            raise IndexError("front from empty deque")
        return self.items[0]

    # See rightmost element
    def back(self):
        if not self.items:
            raise IndexError("back from empty deque")
        return self.items[-1]

    # Check empty
    def empty(self):
        return len(self.items) == 0

    # Size
    def size(self):
        return len(self.items)

    # Print all elements
    def display(self):
        print(self.items)

    # For clean printing
    def __repr__(self):
        return f"Deque({self.items})"
```

```python
d = Deque()

d.append_right(10)
d.append_right(20)
d.append_left(5)

d.display()      # [5, 10, 20]

print(d.pop_left())   # 5
print(d.pop_right())  # 20

print(d.front())      # 10
print(d.back())       # 10
```
