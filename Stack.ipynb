```python
class Stack:
    def __init__(self):
        self.Stack = []

    def push(self, item):
        self.Stack.append(item)

    def pop(self):
        if len(self.Stack) == 0:
            raise IndexError("Empty Stack") # IndexError : Index range out of bound, breaks executiona as it's exception
        else:
            self.Stack.pop() # return self.Stack.pop(), if you want to print

    def top(self):
        return self.Stack[-1] 

    def isempty(self):
        if len(self.Stack) == 0:
            return True
        else:
            return False
        
    def len(self):
        return len(self.Stack)

    def items(self):
        return self.Stack
```

```python
Stack = Stack()
Stack.push(1)
Stack.push(2)
Stack.push(3)
Stack.push(4)

print(Stack.items())

print("Length : ",Stack.len())

Stack.pop()
Stack.pop()

print("Stack Top : ",Stack.top())
print(Stack.items())
```
