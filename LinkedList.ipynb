# LINKED LIST

```python
class Node:
    def __init__(self, data, next = None):
        self.data = data
        self.next = next

class LinkedList:
    def __init__(self,head = None):
        self.head = head

    def insert_front(self, new_node): # if head already enists or doesn't exist case
        temp = self.head
        self.head = new_node
        self.head.next = temp

    def insert_rear(self, new_node): # if LL is empty or not case
        if self.head == None:
            self.head = new_node
        else:
            current = self.head
            while(current.next != None): # current.next selects the last node, current traverses till last node)
                current = current.next
    
            current.next = new_node

    def insert_mid(self, location_data, new_node): #if LL is empty case and not case, location_data doesn't exist
        if self.head == None:
            print("Empty list") # empty = don't insert
            return 
        else:
            current = self.head
            while(current != None and current.data != location_data): # if node doesn't exist exit, if exists check data if not exit loop
                current = current.next
                
            if(current == None):
                print("data not found")
                return

            temp = current.next
            current.next = new_node
            new_node.next = temp

    def __str__(self): # Ignore this for logic, i.e. Traverse LinkedList
        current = self.head
        if self.head == None:
            return "Empty linked list"
        else:
            txt = ""
            current = self.head
            while current != None:
                txt = txt + str(current.data) + "\n"
                current = current.next
        return txt
```

```python
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)
LL = LinkedList()
LL.insert_rear(node1)
LL.insert_rear(node2)
LL.insert_rear(node4)
LL.insert_mid(2,node3)
print(LL)
```
