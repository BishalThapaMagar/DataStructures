```python
class Dequeue:
    def __init__(self):
        self.Dequeue = []

    # REAR
    def add_rear(self, item):
        self.Dequeue.append(item)

    def delete_rear(self):
        if len(self.Dequeue) == 0:
            raise IndexError("Empty Queue")
        return self.Dequeue.pop()

    def look_rear(self):
        return self.Dequeue[-1]

    # FRONT
    def add_front(self, item):
        self.Dequeue.insert(0,item)

    def delete_front(self):
        if len(self.Dequeue) == 0:
            raise IndexError("Empty Queue")
        return self.Dequeue.pop(0)

    def look_front(self):
        return self.Dequeue[0]

    # Basics
    def is_empty(self):
        return len(self.Dequeue) == 0 # if len(self.Dequeue) == 0: return True else: return False

    def len(self):
        return len(self.Dequeue)

    def items(self):
        return self.Dequeue
        
    # Object Print, same as items(self)
    def __str__(self):
        return f"{self.Dequeue}" # if we directly, print(Dequeue())
```

```python
# Example usage
dq = Dequeue()
dq.add_rear(10)
dq.add_front(5)
dq.add_rear(20)
dq.add_front(1)
print(dq)              # Deque([5, 10, 20])
print("Looked rear  : ",dq.look_rear())
print("Deleted rear  : ",dq.delete_rear())    # 5
print("Looked front : ",dq.look_front())
print("Deleted front : ",dq.delete_front()) # 10
print("Length of queue : ",dq.len())  # 20
print(dq)
```
