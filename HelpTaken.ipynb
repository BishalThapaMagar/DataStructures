```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0  # optional, useful for length tracking

    # 1️⃣ Traverse & display
    def traverse(self):
        cur = self.head
        while cur:
            print(cur.data, end=" -> ")
            cur = cur.next
        print("None")

    # 2️⃣ Insert at front
    def insert_front(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        self.size += 1

    # 3️⃣ Insert at back
    def insert_back(self, data):
        new_node = Node(data)
        if not self.head:  # empty list
            self.head = new_node
        else:
            cur = self.head
            while cur.next:
                cur = cur.next
            cur.next = new_node
        self.size += 1

    # 4️⃣ Insert after a given value
    def insert_after(self, target, data):
        cur = self.head
        while cur and cur.data != target:
            cur = cur.next

        if not cur:
            print("Target value not found")
            return

        new_node = Node(data)
        new_node.next = cur.next
        cur.next = new_node
        self.size += 1

    # 5️⃣ Delete from front
    def delete_front(self):
        if not self.head:
            return
        self.head = self.head.next
        self.size -= 1

    # 6️⃣ Delete from back
    def delete_back(self):
        if not self.head:
            return
        if not self.head.next:
            self.head = None
            self.size = 0
            return

        prev = self.head
        cur = self.head.next
        while cur.next:
            prev = cur
            cur = cur.next

        prev.next = None
        self.size -= 1

    # 7️⃣ Delete by value
    def delete_value(self, target):
        if not self.head:
            return

        if self.head.data == target:
            self.head = self.head.next
            self.size -= 1
            return

        prev = self.head
        cur = self.head.next
        while cur and cur.data != target:
            prev = cur
            cur = cur.next

        if cur:
            prev.next = cur.next
            self.size -= 1
        else:
            print("Value not found")

    # 8️⃣ (Optional) Search a value
    def search(self, target):
        cur = self.head
        idx = 0
        while cur:
            if cur.data == target:
                return idx
            cur = cur.next
            idx += 1
        return -1  # not found

    # 9️⃣ Get size
    def get_size(self):
        return self.size
```

```python
# ---------------- TREE IMPLEMENTATION ----------------

class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


class BinaryTree:
    def __init__(self):
        self.root = None

    # Insert in level order (not BST)
    def insert(self, data):
        new_node = TreeNode(data)

        if self.root is None:
            self.root = new_node
            return

        queue = [self.root]

        while queue:
            cur = queue.pop(0)

            if cur.left is None:
                cur.left = new_node
                return
            else:
                queue.append(cur.left)

            if cur.right is None:
                cur.right = new_node
                return
            else:
                queue.append(cur.right)

    # Inorder traversal: Left -> Root -> Right
    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.data, end=" ")
            self.inorder(node.right)

    # Preorder traversal: Root -> Left -> Right
    def preorder(self, node):
        if node:
            print(node.data, end=" ")
            self.preorder(node.left)
            self.preorder(node.right)

    # Postorder traversal: Left -> Right -> Root
    def postorder(self, node):
        if node:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.data, end=" ")


# -------- TREE TEST --------
if __name__ == "__main__":
    bt = BinaryTree()

    for val in [1, 2, 3, 4, 5]:
        bt.insert(val)

    print("Inorder:")
    bt.inorder(bt.root)

    print("\nPreorder:")
    bt.preorder(bt.root)

    print("\nPostorder:")
    bt.postorder(bt.root)
```

```python
# ---------------- GRAPH IMPLEMENTATION ----------------

class Graph:
    def __init__(self):
        self.adj = {}

    # Add edge (undirected)
    def add_edge(self, u, v):
        if u not in self.adj:
            self.adj[u] = []
        if v not in self.adj:
            self.adj[v] = []

        self.adj[u].append(v)
        self.adj[v].append(u)

    # Depth First Search (DFS)
    def dfs(self, node, visited):
        if node in visited:
            return

        print(node, end=" ")
        visited.add(node)

        for neighbor in self.adj.get(node, []):
            self.dfs(neighbor, visited)

    # Breadth First Search (BFS)
    def bfs(self, start):
        visited = set()
        queue = [start]
        visited.add(start)

        while queue:
            node = queue.pop(0)
            print(node, end=" ")

            for neighbor in self.adj.get(node, []):
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)


# -------- GRAPH TEST --------
if __name__ == "__main__":
    g = Graph()

    g.add_edge(0, 1)
    g.add_edge(0, 2)
    g.add_edge(1, 2)
    g.add_edge(2, 3)

    print("DFS:")
    visited = set()
    g.dfs(0, visited)

    print("\nBFS:")
    g.bfs(0)
```

```python
# ---------------- HEAP IMPLEMENTATION ----------------

class MinHeap:
    def __init__(self):
        self.heap = []

    # Insert an element
    def insert(self, val):
        self.heap.append(val)
        self._heapify_up(len(self.heap) - 1)

    # Heapify upward
    def _heapify_up(self, index):
        while index > 0:
            parent = (index - 1) // 2

            if self.heap[index] < self.heap[parent]:
                self.heap[index], self.heap[parent] = self.heap[parent], self.heap[index]
                index = parent
            else:
                break

    # Remove minimum element (root)
    def extract_min(self):
        if not self.heap:
            return None

        if len(self.heap) == 1:
            return self.heap.pop()

        root = self.heap[0]
        self.heap[0] = self.heap.pop()
        self._heapify_down(0)
        return root

    # Heapify downward
    def _heapify_down(self, index):
        size = len(self.heap)

        while True:
            smallest = index
            left = 2 * index + 1
            right = 2 * index + 2

            if left < size and self.heap[left] < self.heap[smallest]:
                smallest = left

            if right < size and self.heap[right] < self.heap[smallest]:
                smallest = right

            if smallest != index:
                self.heap[index], self.heap[smallest] = self.heap[smallest], self.heap[index]
                index = smallest
            else:
                break

    # Display heap
    def display(self):
        print(self.heap)


# -------- HEAP TEST --------
if __name__ == "__main__":
    h = MinHeap()

    h.insert(10)
    h.insert(5)
    h.insert(20)
    h.insert(2)
    h.insert(8)

    h.display()

    print("Extract min:", h.extract_min())
    h.display()

    print("Extract min:", h.extract_min())
    h.display()
```

```python
# ---------------- TRIE IMPLEMENTATION ----------------

class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False


class Trie:
    def __init__(self):
        self.root = TrieNode()

    # Insert a word
    def insert(self, word):
        cur = self.root

        for ch in word:
            if ch not in cur.children:
                cur.children[ch] = TrieNode()
            cur = cur.children[ch]

        cur.is_end = True

    # Search full word
    def search(self, word):
        cur = self.root

        for ch in word:
            if ch not in cur.children:
                return False
            cur = cur.children[ch]

        return cur.is_end

    # Check prefix
    def starts_with(self, prefix):
        cur = self.root

        for ch in prefix:
            if ch not in cur.children:
                return False
            cur = cur.children[ch]

        return True


# -------- TRIE TEST --------
if __name__ == "__main__":
    t = Trie()

    t.insert("apple")
    t.insert("app")
    t.insert("bat")

    print(t.search("app"))      # True
    print(t.search("apple"))    # True
    print(t.search("appl"))     # False

    print(t.starts_with("ap"))  # True
    print(t.starts_with("ba"))  # True
    print(t.starts_with("cat")) # False
```
